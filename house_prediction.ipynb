```python
import pandas as pd
import random as rnd
def effect_fill_na(data, column):
    sm_na = data[column].isna().sum()
    data['qodirjon'] = [x for x in range(len(data))]
    data = data.sort_values(by=[column])
    arra = data[column].dropna()
    real = []
    [real.append(x) for x in arra]
    rand_data = []
    for i in range(sm_na):
        rand_data.append(rnd.choice(real))
    final_data = real + rand_data
    data[column] = final_data
    data = data.sort_values(by=['qodirjon'])
    data.drop('qodirjon', axis=1, inplace=True) 
    return data
```

```python
df = effect_fill_na(df, 'Area')
```
